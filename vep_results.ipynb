```python
import pandas as pd
import numpy as np
```

```python
# function to search for specific name in column
def search_col(df, inp):
    return [i for i in df.columns if inp in i.lower()]
```

```python
# function to make spliceAI prediction scores into floats
def make_float(df):
    for i in ['AG', 'AL', 'DG', 'DL']:
        df[('SpliceAI_pred_DS_'+i)]=df[('SpliceAI_pred_DS_'+i)].astype(float)
    return df
```

```python
vep = pd.read_table('02_output_analysis/first_splice_vars_vep.txt')
all_vars = pd.read_pickle('00_dataframes/all_vars')
```

```python
per_gene = vep.groupby('#Uploaded_variation')
vars_lst = [(var, df) for var,df in per_gene]
```

```python
# right transcript (like in all_vars)
right_transcr = vep[vep['Feature'].str.split('.', expand=True)[0].isin(all_vars['RefSeq accession'].unique())
                   ].reset_index(drop=True)
```

```python
# dictionary for each variant's 'Extra' column
extr_dict = {var:{cat.split('=')[0]:cat.split('=')[1] for cat in right_transcr.iloc[i]['Extra'].split(';')} 
             for i,var in enumerate(right_transcr['#Uploaded_variation'])}
```

```python
# transform dictionary into dataframe, but has to be transposed, then merge with original df, drop 'Extra'
new_df = pd.merge(right_transcr, pd.DataFrame(extr_dict).transpose(), left_on='#Uploaded_variation', 
                  right_on=pd.DataFrame(extr_dict).transpose().index).drop(columns='Extra')
```
